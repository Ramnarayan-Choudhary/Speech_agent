# Local Inference Demo for Indic ASR

This notebook shows how to load a fine-tuned model and transcribe audio locally.

Supported languages: Marathi (mr), Gujarati (gu), Hindi (hi)

```python
from transformers import pipeline

# Load model path
LANGUAGE = "mr"  # Change to "gu" or "hi"
MODEL_PATH = f"/tmp/Speech_agent/models/whisper-{LANGUAGE}-finetuned"

print("Loading oracle model from:", MODEL_PATH)

asr = pipeline(
    "automatic-speech-recognition",
    model=MODEL_PATH,
    device=0 if __import__('torch').cuda.is_available() else -1,
)

# Example audio file path (replace with a real file)
audio_path = "/tmp/Speech_agent/example_data/custom/marathi/clips/sample.wav"

print("Running inference...")
result = asr(audio_path)
print("Transcription:", result["text"])
```
